1. 考虑初始向量 $ v=(1, 0) $ ，给定二维矩阵 $$R(\theta) = \begin{pmatrix} \cos\theta &\sin\theta \\ -\sin\theta & \cos\theta \end{pmatrix}$$，
利用 Numpy 程序实现矩阵乘向量的计算 $v'=R(\theta)v$。通过可视化找到这一向量变化的几何意义。

```python
import numpy as np
import matplotlib.pyplot as plt
from math import cos,sin,pi

theta = 1
v = np.array([1, 0])
R = np.array([[cos(theta), sin(theta)],
              [-sin(theta), cos(theta)]])

v1 = v @ R

print("v =",v)
print("R =",R)
print("v' =",v1)

plt.figure(figsize=(3, 3))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v')
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label="v'")
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid(True)
plt.title('v @ R(theta)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

借助上图的可视化，不难发现该变化实际上就是将向量 $v$ 逆时针方向旋转了 $\theta$ 角。同时这一旋转与 $v$ 无关。
